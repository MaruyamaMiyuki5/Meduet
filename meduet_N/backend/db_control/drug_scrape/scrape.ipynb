```python
url = 'https://www.pmda.go.jp/PmdaSearch/otcDetail/170039_J0601006247_04_01'
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
hds = []
elems = soup.select('td.head')
for elem in elems:
  txt = elem.text.replace('\n','').replace('\t','')
  hds.append(txt)

bds = []
elems = soup.select('td.deta')
for elem in elems:
  txt = elem.text.replace('\n','').replace('\t','')
  bds.append(txt)

```

```python
df = pd.DataFrame()
for i  in range(len(hds)):
  df[f'{hds[i]}'] = [bds[i]]
df
```
